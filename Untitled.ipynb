```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 2*np.pi, 100)
plt.plot(np.cos(t), np.sin(t), 'b-', label='Primary Orbit')
plt.plot(0.6*np.cos(3*t), 0.6*np.sin(3*t), 'r--', label='Satellite Orbit')
plt.axis('equal')
plt.grid(True)
plt.title('Satellite Orbit')
plt.legend()
plt.show()
```
